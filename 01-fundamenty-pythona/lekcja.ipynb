## 4. Operatory arytmetyczne

Python obsuguje wszystkie standardowe operacje matematyczne:

````python
a = 15
b = 4

wynik_dodwania = a + b
wynik_odejmowania = a - b
wynik_potegowania = a ** b

print(f"Dodawanie: {a} + {b} = {wynik_dodwania}")
print(f"Odejmowanie: {a} - {b} = {wynik_odejmowania}")
print(f"Mno偶enie: {a} * {b} = {a * b}")
print(f"Dzielenie: {a} / {b} = {a / b}")  # zawsze zwraca float
print(f"Dzielenie cakowite: {a} // {b} = {a // b}")  # zaokrgla w d贸
print(f"Reszta z dzielenia: {a} % {b} = {a % b}")  # modulo
print(f"Potgowanie: {a} ** {b} = {wynik_potegowania}")  # 15 do potgi 4
````

Zamiast pisa `x = x + 5`, mo偶emy u偶y skr贸conej wersji:

````python
liczba = 10

liczba += 5   # to samo co: liczba = liczba + 5
print(f"Po dodaniu 5: {liczba}")

liczba -= 3   # to samo co: liczba = liczba - 3
print(f"Po odjciu 3: {liczba}")

liczba *= 2   # to samo co: licznik = licznik * 2
print(f"Po pomno偶eniu przez 2: {liczba}")

liczba //= 4  # to samo co: licznik = licznik // 4
print(f"Po dzieleniu cakowitym przez 4: {liczba}")
````

Operatory por贸wnania zwracaj warto `True` lub `False`:

Su偶 do czenia warunk贸w logicznych:

### Konwersja midzy typami

Czasem musisz przekonwertowa dane z jednego typu na inny:

````python
# String na liczb
tekst_liczba = "42"
liczba = int(tekst_liczba)
print(f"int: {liczba}, typ: {type(liczba)}")

# String na float
tekst_float = "3.14"
liczba_float = float(tekst_float)
print(f"float: {liczba_float}, typ: {type(liczba_float)}")

# Liczba na string
liczba = 100
tekst = str(liczba)
print(f"str: '{tekst}', typ: {type(tekst)}")

# Konwersja na bool
print(f"bool(1): {bool(1)}")      # True
print(f"bool(0): {bool(0)}")      # False
print(f"bool(''): {bool('')}")    # False (pusty string)
print(f"bool('a'): {bool('a')}")  # True (niepusty string)
````
